```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1000
x = np.random.rand(n)
width = 0.05
histmin = np.floor(min(x))
histmax = np.ceil(max(x))+width

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=100,alpha=0.75,edgecolor="black",color="red")
plt.ylabel("N per bin")
plt.xlabel("x")
plt.savefig("exercise9.pdf",bbox_inches="tight",facecolor="white")
```
