```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.gamma(1, 100, 1000)


width = 0.05
histmin = np.floor(min(x))
histmax = np.ceil(max(x))+width


bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=100,alpha=0.75,edgecolor="black",color="red")
plt.ylabel("N per bin")
plt.xlabel("x")
plt.savefig("Project10.pdf",bbox_inches="tight",facecolor="white")
```
